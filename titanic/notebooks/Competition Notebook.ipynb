```python
import pandas as pd
import numpy as np
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("../../datasets/titanic/train.csv")
```

```python
df.head()
```

```python
test_df = pd.read_csv("../../datasets/titanic/test.csv")
test_df.head()
```

```python
women = df.loc[df.Sex=="female"]["Survived"]
rate_women = sum(women)/len(women)
print("Survival rate of women: {}".format(rate_women))
```

```python
men = df.loc[df.Sex=="male"]["Survived"]
rate_men = sum(men)/len(men)
print("Survival rate of women: {}".format(rate_men))
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch"]
X, y = pd.get_dummies(df[features]), df["Survived"]
X_test = pd.get_dummies(test_df[features])
```

```python
# Initialize the model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

# Train the model
model.fit(X, y)

# Predict using the model
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': predictions})
output.to_csv("titanic_051320_1.csv", index=False)
```

### Let's think about this a bit.

In the data available, the feature 'Survived' is what we're trying to predict, and everything else are potential input features that can help us predict the outcome.


We can reasonably assume that the name of a person had nothing to do with their chance of survival. 

```python
df.sort_values('Cabin')
```

```python
df['cabin_letter'] = df.Cabin.str[0]
```

```python
df.sort_values('cabin_letter')
```

```python
print(df.loc[df.cabin_letter == "A"].count(), df.loc[df.cabin_letter == "A", 'Survived'].sum(axis = 0))
```

```python
cabin_letters = df.cabin_letter.unique()
```

```python
cabin_letters
```

```python
len
```

```python
survival = []

for letter in cabin_letters:
    if letter == letter: # check will fail on NaN value
        a, b, c = df.loc[df.cabin_letter==letter, 'Survived'].sum(), len(df.loc[df.cabin_letter==letter]), (df.loc[df.cabin_letter==letter, 'Survived'].sum())/len(df.loc[df.cabin_letter==letter]) 
        
    else:
        a, b, c = df.loc[df.cabin_letter.isnull(), 'Survived'].sum(), len(df.loc[df.cabin_letter.isnull()]), df.loc[df.cabin_letter.isnull(), 'Survived'].sum()/len(df.loc[df.cabin_letter.isnull()])
    
    survival.append((letter, a, b, c))
        
```

```python
survival
```
